## 13) Write a function to compute the argmin of a numpy array. The function should take a numpy array as input and return the index of the minimum element. You can use the np.argmin function to verify your solution.

```python
## An Example taken for showing the asked result

matrix = np.array([[[1,4],[2,3]],
          [[2,2],[4,3]],
         [[0,0],[0,0]]])

print("Argmin obtained from inbuilt function is",np.argmin(matrix))


def findargmin(matrix):
    matrix_flattened = matrix.flatten()
    min_element = np.min(matrix_flattened)
    index = np.where(matrix_flattened == min_element)
    first_occurences = index[0][0]
    return first_occurences

print("Argmin obtained using userdefined function",findargmin(matrix))
```

## 3) Finding the highest element in a list requires one pass of the array. Finding the second highest element requires 2 passes of the the array. Using this method, what is the time complexity of finding the median of the array? Can you suggest a better method? Can you implement both these methods in Python and compare against numpy.median routine in terms of time?

For finding the nth highest element n passes are required. Therefore for each pass we need $T(n) = O(n)$
To find median median, we need to find $n/2^{th}$ highest element, therefore time taken = $T(n) = O(n \times n/2)$

We can use sorting to find the median with a time complexity of $T(n) = O(n \space log \space n)$

```python
## By using the mentioned method
arr = np.random.randint(0,100,10)
print(f'Median obtained using np.median method is {np.median(arr)}')

def findMedian1(arr):
    n = len(arr)
    for i in range(len(arr)//2):
        maxelement = np.max(arr)
        arr = np.delete(arr,arr.argmax())
    if(n % 2 == 0):
        return (np.max(arr) + maxelement)/2
    else:
        return np.mex()
        
print(f'Median obtained using the findMedian1 method is {findMedian1(arr)}')
```

```python
## by using the better method
arr1 = np.random.randint(0,100,9)
print(f'Median obtained using np.median method is {np.median(arr1)}')

def findMedian2(arr1):
    sorted_arr = np.sort(arr1)
    n = arr1.shape[0]
    if n % 2:
        return sorted_arr[(n)//2]
    else:
        return (sorted_arr[(n-1)//2] + sorted_arr[(n)//2])/2

print(f'Median obtained using the findMedian2 method is {findMedian2(arr1)}')
```
